## CT image
Let's build an interactive figure that goes through the middle slices of the first ten .nii.gz CT files. This may take some time.

### AbdomenCT-1K dataset

````python
# other dataset
from ipywidgets import interact, IntSlider, fixed
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch

# Load all CT images from the directory
ct_dir = r"D:/yao/Rice University/Academic/Introduction to Deep Learning/final project/Image/Image/"
mask_dir = r"D:/yao/Rice University/Academic/Introduction to Deep Learning/final project/Mask/Mask/"
label_dict = {0: 'background', 1: 'liver', 2: 'kidney', 3: 'spleen', 4: 'pancreas'}

ct_files = sorted(glob.glob(os.path.join(ct_dir, "*.nii.gz")))
mask_files = sorted(glob.glob(os.path.join(mask_dir, "*.nii.gz")))

if not ct_files:
    print("No .nii or .nii.gz files found!")
else:
    if not mask_files:
        print("No .nii or .nii.gz files found!")
    else:
        print(f"Found {len(ct_files)} CT images")
        print(f"Found {len(mask_files)} Mask images")
        count = 0

        # Load all middle slices for both CT and masks
        ct_middle_slices = []
        mask_middle_slices = []
        filenames = []

        for ct_file, mask_file in zip(ct_files, mask_files):
            ct_img = nib.load(ct_file)
            ct_data = ct_img.get_fdata()
            middle_slice_idx = ct_data.shape[2] // 2
            ct_middle_slices.append(ct_data[:, :, middle_slice_idx])
            
            mask_img = nib.load(mask_file)
            mask_data = mask_img.get_fdata()
            mask_middle_slices.append(mask_data[:, :, middle_slice_idx])
            
            filenames.append(os.path.basename(ct_file))
            count += 1
            if count >= 10:
                break

        # Initialize with the first image
        ct_middle_slices = np.array(ct_middle_slices)
        mask_middle_slices = np.array(mask_middle_slices)


        def plot_slice_abdomen(current_image, ct_slices, mask_slices):
            plt.close('all')  # Close any existing figures
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
            
            # Plot CT image
            im1 = ax1.imshow(ct_slices[current_image], cmap='gray')
            ax1.set_title(f'CT Image {current_image + 1}/{len(ct_slices)}: {filenames[current_image]}')
            ax1.axis('off')
            cbar1 = plt.colorbar(im1, ax=ax1, shrink=0.7)
            cbar1.set_label('Pixel Value (HU)')
            
            # Plot mask with custom colormap - fixed range for all slices
            mask_slice = mask_slices[current_image]
            
            im2 = ax2.imshow(mask_slice, cmap='tab20', vmin=0, vmax=len(label_dict)-1)
            ax2.set_title(f'Label Mask {current_image + 1}/{len(mask_slices)}')
            ax2.axis('off')
            
            # Create legend with ALL labels in the dictionary
            legend_elements = [Patch(facecolor=plt.cm.tab20(label/(len(label_dict)-1)), 
                                    label=f'{label}: {label_dict[label]}') 
                              for label in sorted(label_dict.keys())]
            ax2.legend(handles=legend_elements, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=8)
            
            plt.tight_layout()

        interact(plot_slice_abdomen,
                  current_image=IntSlider(min=0, max=count - 1, step=1, value=(count-1)//2, description='Slice Index'),
                  ct_slices=fixed(ct_middle_slices),
                  mask_slices=fixed(mask_middle_slices)
                 )
````

Among these various matadata, qoffset is just a translation, setting the position of the origin (i=0,j=0,k=0) in real space.

One thing to notice is the physical size for each pixel of these .nii.gz CT image, which is stored in pixdim[1:4]. For example, some may have $\Delta z=4mm$ but the others may have $\Delta z=1mm$. This means that one pixel corresponds to different real-space physical volume. However, it seems that we don't need to care about $\Delta z$, since we only consider 2D images. Different $\Delta x,\Delta y$ also don't matter, since it in fact acts as scaling, saving the step of performing this data augmentation process.

Below shows the **histogram of physical pixel sizes** among the 'other' dataset and pku dataset.

````python
import matplotlib.pyplot as plt
def plot_histogram(folder_path, dataset_name=None):
    x_spacings = []
    y_spacings = []
    z_spacings = []
    if os.path.exists(folder_path):
        nii_files = sorted([f for f in os.listdir(folder_path) if f.endswith('.nii.gz')])
        print(f"Found {len(nii_files)} files. collecting metadata...")

        for f in nii_files:
            file_path = os.path.join(folder_path, f)
            try:
                img = nib.load(file_path)
                header = img.header
                # pixdim[1], [2], [3] correspond to spatial dimensions x, y, z
                x_spacings.append(header['pixdim'][1])
                y_spacings.append(header['pixdim'][2])
                z_spacings.append(header['pixdim'][3])
            except Exception as e:
                print(f"Error reading {f}: {e}")

        # Plotting
        fig, axes = plt.subplots(1, 3, figsize=(18, 5))

        axes[0].hist(x_spacings, bins=20, color='skyblue', edgecolor='black')
        axes[0].set_title(f'Histogram of Pixel Size (Dim 1) in {dataset_name} dataset')
        axes[0].set_xlabel('Size (mm)')
        axes[0].set_ylabel('Count')

        axes[1].hist(y_spacings, bins=20, color='lightgreen', edgecolor='black')
        axes[1].set_title(f'Histogram of Pixel Size (Dim 2) in {dataset_name} dataset')
        axes[1].set_xlabel('Size (mm)')

        axes[2].hist(z_spacings, bins=20, color='salmon', edgecolor='black')
        axes[2].set_title(f'Histogram of Pixel Size (Dim 3) in {dataset_name} dataset')
        axes[2].set_xlabel('Size (mm)')

        plt.tight_layout()
        plt.show()

    else:
        print(f"Directory not found: {folder_path}")

# for other dataset
folder_path = "D:/yao/Rice University/Academic/Introduction to Deep Learning/final project/Image/Image/"
plot_histogram(folder_path, dataset_name="AbdomenCT-1K")
# For pku dataset
folder_path = "D:/yao/PkU/Academic2024-2025/medical_image_segmentation/pku_train_dataset/ct/" 
plot_histogram(folder_path, dataset_name="pku")
````

# Frequency of labels
Not every label appears equally likely throughout these datasets. I've plotted another histogram showing the frequency of each label for the two dataset. 

````python
def label_histogram(label_path, dataset_name=None):
    label_values = {}
    if os.path.exists(label_path):
        nii_files = sorted([f for f in os.listdir(label_path) if f.endswith('.nii.gz')])
        print(f"Found {len(nii_files)} files. collecting label values...")

        for f in nii_files:
            file_path = os.path.join(label_path, f)
            try:
                img = nib.load(file_path)
                data = img.get_fdata().astype(int)
                unique_vals = np.unique(data).tolist()
                for val in unique_vals:
                    if val==0:
                        continue
                    if val in label_values:
                        label_values[val] += np.sum(data == val)
                    else:
                        label_values[val] = np.sum(data == val)
            except Exception as e:
                print(f"Error reading {f}: {e}")

        # Plotting
        plt.figure(figsize=(10, 5))
        plt.bar(label_values.keys(), label_values.values(), color='violet', edgecolor='black')
        plt.title(f'Histogram of Labels in {dataset_name} dataset')
        plt.xlabel('Label Value')
        plt.ylabel('Count')
        plt.xticks(range(min(label_values), max(label_values) + 1))
        plt.show()

    else:
        print(f"Directory not found: {label_path}")

# for other dataset
label_path = "D:/yao/Rice University/Academic/Introduction to Deep Learning/final project/Mask/Mask/"
label_histogram(label_path, dataset_name="AbdomenCT-1K")
# for pku dataset
label_path = "D:/yao/PkU/Academic2024-2025/medical_image_segmentation/pku_train_dataset/label/"
label_histogram(label_path, dataset_name="pku")
````

# Organ HU range
Values in CT images are in units of HU, usually ranging from -1000 to 1000, but the HU values of abdominal organs have smaller ranges. Let's see the HU value distribution for each organ.

````python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
def HU_distribution(image_path, label_path, label_dictionary, dataset_name=None):
    # label_dictionary looks like: {0: 'background', 1: 'liver', 2: 'kidney', 3: 'spleen', 4: 'pancreas'}
    # And the keys contain all possible labels in the dataset, with no redundancy.
    
    # Check if paths exist
    if not os.path.exists(image_path) or not os.path.exists(label_path):
        print(f"Directory not found: {image_path} or {label_path}")
        return
    
    # Get sorted file lists
    image_files = sorted([f for f in os.listdir(image_path) if f.endswith('.nii.gz')])
    label_files = sorted([f for f in os.listdir(label_path) if f.endswith('.nii.gz')])
    
    print(f"Processing {len(image_files)} CT images for {dataset_name} dataset...")
    
    # Dictionary to store HU values for each label (skip background to save memory)
    organ_hu_values = {label: [] for label in label_dictionary.keys() if label != 0}
    
    # Process each CT image and corresponding mask
    for img_file, lbl_file in zip(image_files, label_files):
        img_path = os.path.join(image_path, img_file)
        lbl_path = os.path.join(label_path, lbl_file)
        
        try:
            # Load CT image and mask
            ct_img = nib.load(img_path)
            ct_data = ct_img.get_fdata()
            
            mask_img = nib.load(lbl_path)
            mask_data = mask_img.get_fdata().astype(int)
            
            # Extract HU values for each organ (skip background label 0)
            for label in label_dictionary.keys():
                if label == 0:  # Skip background to avoid memory issues
                    continue
                mask_positions = (mask_data == label)
                if np.sum(mask_positions) > 0:  # If organ exists in this slice
                    hu_vals = ct_data[mask_positions]
                    organ_hu_values[label].extend(hu_vals.flatten().tolist())
        
        except Exception as e:
            print(f"Error processing {img_file}: {e}")
    
    # Calculate statistics for each organ
    stats = {}
    for label, hu_vals in organ_hu_values.items():
        if len(hu_vals) > 0:
            stats[label] = {
                'mean': np.mean(hu_vals),
                'min': np.min(hu_vals),
                'max': np.max(hu_vals),
                'p10': np.percentile(hu_vals, 10),
                'p90': np.percentile(hu_vals, 90),
                'count': len(hu_vals)
            }
        else:
            stats[label] = None
    
    # Plotting
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Prepare data for plotting (only organs with stats, excluding background which wasn't processed)
    labels_to_plot = [label for label in sorted(label_dictionary.keys()) if label in stats and stats[label] is not None]
    x_positions = np.arange(len(labels_to_plot))
    
    for i, label in enumerate(labels_to_plot):
        stat = stats[label]
        
        # Draw min-max range as a light gray bar
        bar_height = stat['max'] - stat['min']
        ax.bar(i, bar_height, bottom=stat['min'], width=0.4, 
               color='lightgray', alpha=0.3, edgecolor='gray', linewidth=1.5)
        
        # Draw 10%-90% percentile as error bar (vertical line)
        ax.plot([i, i], [stat['p10'], stat['p90']], 
                color='blue', linewidth=3, label='10%-90%ile' if i == 0 else '')
        
        # Draw mean as a point
        ax.scatter(i, stat['mean'], color='red', s=100, zorder=5, 
                  label='Mean' if i == 0 else '')
        
        # Print statistics
        print(f"Label {label} ({label_dictionary[label]}): "
              f"Mean={stat['mean']:.1f}, Min={stat['min']:.1f}, Max={stat['max']:.1f}, "
              f"10%={stat['p10']:.1f}, 90%={stat['p90']:.1f}, Count={stat['count']}")
    
    # Set x-axis labels
    x_labels = [f"{label}: {label_dictionary[label]}" for label in labels_to_plot]
    ax.set_xticks(x_positions)
    ax.set_xticklabels(x_labels, rotation=45, ha='right')
    
    # Labels and title
    ax.set_ylabel('HU Value', fontsize=12)
    ax.set_xlabel('Organ', fontsize=12)
    ax.set_title(f'HU Value Distribution by Organ - {dataset_name} Dataset', fontsize=14, fontweight='bold')
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    ax.legend(loc='upper right')
    
    plt.tight_layout()
    plt.show()

# Test on AbdomenCT-1K dataset
image_path = "D:/yao/Rice University/Academic/Introduction to Deep Learning/final project/Image/Image/"
label_path = "D:/yao/Rice University/Academic/Introduction to Deep Learning/final project/Mask/Mask/"
label_dict_abdomen = {0: 'background', 1: 'liver', 2: 'kidney', 3: 'spleen', 4: 'pancreas'}
HU_distribution(image_path, label_path, label_dict_abdomen, dataset_name="AbdomenCT-1K")
````
